```python
from pdf2image import convert_from_path
import os


input_folder = "/kaggle/input/datasets/satarupadeb/renaissance/document"
output_folder = "/kaggle/working/images"

os.makedirs(output_folder, exist_ok=True)

img_count = 0

for file in os.listdir(input_folder):
    if file.lower().endswith(".pdf"):
        pdf_path = os.path.join(input_folder, file)
        
        pages = convert_from_path(pdf_path)
        
        for i, page in enumerate(pages):
            img_name = f"{file.replace('.pdf','')}_page_{i+1:03d}.jpg"
            page.save(os.path.join(output_folder, img_name), "JPEG")
            img_count += 1

print(f"Total images created: {img_count}")
```
